```python
ndf=df[['X12','X13']]
ndf.corr()
```

```python
var = np.cov([df.X12, df.X17])
var
```

```python
t=ndf.var().sum()
varX12=var[0][0]
varX13=var[1][1]
```

```python
print('Varianza total \t\t:', t.round(2) )
print('Varianza prop. de X12 \t:', ((varX12/t)*100).round(2),'%' )
print('Varianza prop. de X13 \t:', ((varX13/t)*100).round(2),'%' )
```

```python
import seaborn as sns
```

```python
p = sns.regplot(data = df,  
                x = df.X12, 
                y = df.X13
               ) 
p
```

```python
from sklearn.decomposition import PCA
```

```python
pcs = PCA(n_components=2)

pcs.fit(ndf)

pcsSummary = pd.DataFrame({'Standard deviation': np.sqrt(pcs.explained_variance_),
                          'Proportion of variance': pcs.explained_variance_ratio_,
                          'Cumulative proportion': np.cumsum(pcs.explained_variance_ratio_)#column 
                          }
                          )
pcsSummary
```

```python

pcsSummary = pcsSummary.transpose()
pcsSummary.round(2)
```

PC1 es el 98% de la variabilidad total
PC2 es solo el 2%
Por lo tanto si descartamos PC2 mantenemos el 98%

Scree Plot

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
PC_components = np.arange(pcs.n_components_) + 1
#PC_components

_ = sns.set(style = 'whitegrid', 
            font_scale = 1.2
            )

fig, ax = plt.subplots(figsize=(10, 7))

_ = sns.barplot(x = PC_components, 
                y = pcs.explained_variance_ratio_, 
                color = 'b'
                )

_ = sns.lineplot(x = PC_components-1, 
                 y = np.cumsum(pcs.explained_variance_ratio_), 
                 color = 'black', 
                 linestyle = '-', 
                 linewidth = 2, 
                 marker = 'o', 
                 markersize = 8
                 )

plt.title('Scree Plot')
plt.xlabel('N-th Principal Component')
plt.ylabel('Variance Explained')
plt.ylim(0, 1)
plt.show()
```

```python
pcs.components_.round(4)
```

```python
pcs.components_.transpose()
```

```python
pcsComponents_df = pd.DataFrame(pcs.components_.transpose(), 
                                columns=['PC1', 'PC2'],
                                index=['X12', 'X13']
                                )
pcsComponents_df
```

```python
scores = pd.DataFrame(pcs.transform(df[['X12', 'X13']]),
                      columns=['x', 'y']
                      )
scores.head()
```
