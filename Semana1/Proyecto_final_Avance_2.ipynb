##**Escribe los pasos estándares en un máximo de 3 oraciones, por cada fase**

###CRISP-DM (Cross-Industry Standard Process for Data Mining,) es una metodología probada para orientar sus trabajos de minería de datos que permite llevar a cabo un proceso determinado y alcanzar objetivo definido. CRISP-DM define un ciclo de vida enfocado a la exploración y análisis de los datos que consta de seis fases.



1. **Business understanding.** En este paso se define cual es el objetivo del proyecto considerando las necesidades del negocio. Se debe tener claro el problema que se desea resolver y la solución de ciencia de datos que se abarcara, en esta fase se define detalles como el plan y análisis del costo y beneficio.
2. **Data understanding.** Esta fase inicia con la recolección de datos, la comprensión de su naturaleza, calidad de la información, accesibilidad, su origen si es interna o externa, se plantean las primeras hipótesis sobre el análisis de la información.
3. **Data preparation.** Cubre todas las actividades necesarias para construir el conjunto de datos final, implica la integración de datos, muestreos, transformación, limpieza, construcción, es la fase más larga en el ciclo de vida del proyecto, puede representar hasta el 80% de tiempo.
4. **Modelling.** Se selecciona la técnica adecuada para el modelo de datos, considerando la generación de pruebas, la construcción del modelo y su evaluación.
5. **Evaluation.** En esta fase se valida el modelo por parte de expertos, los resultados empresariales, se definen los procesos de revisión y los próximos pasos a seguir. Es clave identificar la mejora de los procesos que se puede implementar y la aprobación de la continuidad de la implementación.
6. **Deployment.** En esta fase se desarrolla y documenta el plan de despliegue, el plan de monitoreo y mantenimiento para la parte operacional y prevención de fallas, se entrega un reporte final con resultados del procesamiento de datos y se debe producir un informe final del proyecto con la retrospectiva.
